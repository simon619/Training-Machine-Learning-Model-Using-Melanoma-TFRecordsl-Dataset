```python
# Check how many values are missing in the anatom_site_general_challenge column
print("Number of rows with null values in anatom_site_general_challenge column: ", task['anatom_site_general_challenge'].isnull().sum())

# Subset the volunteer dataset

task_subset = task[task['anatom_site_general_challenge'].notnull()]

# Print out the shape of the subset
print("Shape after removing null values: ", task_subset.shape)
```

```python
print("Shape of dataframe before dropping:", task.shape)
task = task.dropna(axis = 0, subset = ['anatom_site_general_challenge'])
print("Shape after dropping:", task.shape)
```

```python
from sklearn.impute import SimpleImputer

impute = SimpleImputer(missing_values=np.nan, strategy='mean')

impute.fit(task[['age_approx']])

task['age_approx'] = impute.transform(task[['age_approx']])
print("Shape after imputing:", task.shape)
```

```python
# Check how many values are missing in the sex column
print("Number of rows with null values in sex column: ", task['sex'].isnull().sum())

# Subset the volunteer dataset

task_subset = task[task['sex'].notnull()]

# Print out the shape of the subset
print("Shape after removing null values: ", task_subset.shape)
```

```python
print("Shape of dataframe before dropping:", task.shape)
task = task.dropna(axis = 0, subset = ['sex'])
print("Shape after dropping:", task.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled_MinMax = scaler.transform(X_train)
X_test_scaled_MinMax = scaler.transform(X_test)
```

**KNeighborsClassifier**

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier()
knn.fit(X_train_scaled_MinMax, y_train)

# scoring on the scaled test set
print("KNN test accuracy: {:.2f}".format(knn.score(X_test_scaled_MinMax, y_test)))
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled_standard = scaler.transform(X_train)
X_test_scaled_standard = scaler.transform(X_test)
```

```python
X_train_scaled_standard
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier()
knn.fit(X_train_scaled_standard, y_train)

# scoring on the scaled test set
knn_score = knn.score(X_test_scaled_standard, y_test)
name.append('KNN')
print("KNN test accuracy: {:.2f}".format(knn_score))
model_score.append(knn_score)
# model_pred.append(knn.prediction(X_test_scaled_standard))
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

#Train the model
log_model = LogisticRegression()
log_model.fit(X_train_scaled_standard, y_train) #Training the model
name.append('Logistic Regression')
log_predictions = log_model.predict(X_test_scaled_standard)
print(log_predictions)# printing predictions
```

```python
print("LogisticRegression test accuracy: {:.2f}".format(accuracy_score(y_test, log_predictions)))
```

```python
log_score = log_model.score(X_test_scaled_standard, y_test)
print("LogisticRegression test accuracy: {:.2f}".format(log_model.score(X_test_scaled_standard, y_test)))
model_pred.append(log_model.predict(X_test_scaled_standard))
model_score.append(log_score)
```

**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(criterion='entropy',random_state=1)
tree_model.fit(X_train_scaled_standard, y_train)
name.append('Decision Tree')
tree_prediction = tree_model.predict(X_test_scaled_standard)
score=accuracy_score(tree_prediction, y_test)
print(score)
```

```python
tree_score = tree_model.score(X_test_scaled_standard, y_test)
print("Decision Tree test accuracy: {:.2f}".format(tree_score))
model_pred.append(tree_model.predict(X_test_scaled_standard))
model_score.append(tree_score)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(tree_model,
               feature_names = ['sex', 'age_approx', 'anatom_site_general_challenge', 'diagnosis', 'target', 'tfrecord', 'width', 'height'], 
               class_names = ['benign', 'malignant'],
               filled = True);
```
